```python
import spacy
from spacy import displacy
import re
nlp = spacy.load('en_core_web_trf')
```

```python
import yaml

with open('../data/pg35041-paragraphs.yaml', 'r', encoding='utf-8') as f:
    paragraphs = yaml.safe_load(f.read())  
```

```python
doc = nlp(paragraphs[36])
displacy.render(doc, style="ent", jupyter=True)
```

## Listing up all the organizations in the book

```python
orgs = []
for doc in nlp.pipe(paragraphs):
    for ent in doc.ents:
        if ent.label_ == 'ORG':
            orgs.append(ent.text)
orgs
```

```python
import collections

orgs_freq = collections.Counter(orgs)
orgs_freq
```
